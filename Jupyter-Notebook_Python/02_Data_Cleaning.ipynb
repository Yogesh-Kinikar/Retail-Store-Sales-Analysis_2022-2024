```python
# Import all necessary libraries
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt     
```

```python
store_data = pd.read_csv(r"C:\Users\kiyo3001\Desktop\Personal\Python\Store_Data\retail_store_sales.csv")  #Load the dataset
store_data.head()    # Display First Five Rows
```

```python

store_data.columns = (
    store_data.columns
    .str.strip()                     # Remove Spaces
    .str.lower()                     # Convert To Lowercase
    .str.replace(' ','_')            # Replace Spaces With Underscores
)
print(store_data.columns)            # Display The Column Names
```

```python
store_data.shape                    # Display The Shape Of The Dataset
```

```python
store_data.info()                   # Display The Info Of The Dataset
```

```python
store_data.duplicated().sum()        # Check For Duplicated Rows
```

```python
store_data.isna().sum()           # Check For Missing Values
```

```python
store_data['item']= store_data['item'].fillna('Unknown Item')    # Fill Missing Values In 'item' Column With 'Unknown Item'


medians= store_data.groupby('item')[['price_per_unit','quantity']].median()        # Calculate Median Of 'price_per_unit' And 'quantity' Grouped By 'item'
for col in ['price_per_unit','quantity']:
    store_data[col]= store_data[col].fillna(store_data['item'].map(medians[col]))   # Fill Missing Values In 'price_per_unit' And 'quantity' Columns With Corresponding Median Values

store_data['total_spent']= store_data['price_per_unit'] * store_data['quantity']    # compute 'total_spent' By Multiplying 'price_per_unit' And 'quantity'

store_data['discount_applied'] = store_data['discount_applied'].fillna(store_data['discount_applied'].mode()[0])  # Fill Missing Values In 'discount_applied' Column With Mode)
```

```python
store_data.isna().sum()           # Check For Missing Values
```

```python
store_data['transaction_date'] = pd.to_datetime(store_data['transaction_date'])  # Convert 'transaction_date' To Datetime Format
```
