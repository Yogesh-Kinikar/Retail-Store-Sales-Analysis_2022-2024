
# Introduction -  Retail Store Data Sales Analysis
This project analyzes transaction data from an online store. The dataset includes:
- Transaction details (Transaction ID, date, customer ID)
- Product information (category, item, price per unit)
- Quantity and total spending
- Payment method and discount applied

**Objectives:**
- Understand sales trends over time
- Segment customers based on purchasing behavior
- Identify top contributors to revenue using Pareto analysis
