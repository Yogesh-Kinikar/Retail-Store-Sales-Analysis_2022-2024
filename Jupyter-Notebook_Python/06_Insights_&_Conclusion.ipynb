## Insights & Conclusion


### 1. Executive Summary
The analysis of 12,575 retail transactions reveals strong seasonal trends, high-value customer segments, and category-level performance insights. This report summarizes key findings and actionable recommendations.







### 2. Key Insights

### **Sales Trends**
- **Yearly:** 2024 had the highest sales (~554K), Followed by 2022(~539K) and 2023(~515K)
- **Monthly:** Sales fluctuate between 40K–55K, with peaks in December - February.

### **Category Performance**
- **Top Categories:** Butchers (217K), Electric Household Essentials (213K), Furniture (205K).
- **Least Performing:** Milk Products (190K).

### **Customer Analysis**
- **Top Spenders:** CUST_24 (71K), CUST_08 (70K).
- **Most Frequent Buyers:** CUST_05 (544 transactions), CUST_24 (543 transactions).
- **Average Spending:** ~\$130 per transaction for frequent buyers.
- **Need to Focus on these Customers since they are both Frequent customers and top buyers:** CUST_24, CUST_08, CUST_05, CUST_13, CUST_23

### **Payment Methods**
- Cash (34.3%), Digital Wallet (33%), Credit Card (32.8%) – fairly balanced distribution.

### **Discount Impact**
- Discounts slightly reduce average spending per transaction, suggesting they drive quantity rather than value.

### **Correlation Analysis**
- Strong correlation between **quantity** and **total_spent** (0.71).
- Moderate correlation between **price_per_unit** and **total_spent** (0.63).

## 3. Business Recommendations
- **Focus on High-Value Customers:** Implement loyalty programs for top spenders/Frequent buyers
- **Seasonal Promotions:** Target December to February for major campaigns.
- **Category Strategy:** Increase marketing for Butchers and Electric Essentials.
- **Digital Wallet Adoption:** Promote wallet-based offers to leverage growing trend.
- **Discount Optimization:** Use discounts strategically to boost volume without eroding margins.

## 4. Future Scope
- Predictive modeling for **sales forecasting**.
- Customer segmentation using **clustering algorithms**.
- Personalized recommendation engine based on purchase history.
