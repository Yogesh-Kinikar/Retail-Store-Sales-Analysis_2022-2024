### Customer Segmentation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\kiyo3001\Desktop\Personal\Python\csv\cleaned_retail_sales.csv")
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df.head()
```

```python

# 1. Most frequent customers with average spending and top category (Regular Customers)
customer_stats = df.groupby('customer_id').agg(
    Frequency=('transaction_id', 'count'),
    Avg_Spending=('total_spent', 'mean')
).reset_index()

# Find most purchased category for each customer
most_category = df.groupby(['customer_id', 'category']).size().reset_index(name='count')
most_category = most_category.sort_values(['customer_id', 'count'], ascending=[True, False])
most_category = most_category.groupby('customer_id').first().reset_index()

# Merge stats with category
customer_stats = customer_stats.merge(most_category[['customer_id', 'category']], on='customer_id')

# Sort by frequency
most_frequent_customers = customer_stats.sort_values(by='Frequency', ascending=False).head(10)
print("Most Frequent Customers:")
print(most_frequent_customers)


```

```python
# 2. Top 10 customers by highest total sales with top category (VIP customers)
top_customers = df.groupby('customer_id').agg(
    Total_Sales=('total_spent', 'sum')
).reset_index()

top_customers = top_customers.merge(most_category[['customer_id', 'category']], on='customer_id')
top_10_customers = top_customers.sort_values(by='Total_Sales', ascending=False).head(10)
print("\nTop 10 Customers by Total Sales:")
print(top_10_customers)
```
